```python
%%markdown
## Strategy : 3
### Fill The Values With Scikit-Learn
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train3 = train.copy()
```

```python
Mis_Str_var3 = [x for x in train3.select_dtypes(include = ['object']).keys() if train3[x].isnull().sum()>0]
Mis_Num_var3 = [x for x in train3.select_dtypes(include = ['int64', 'float64']).keys() if train3[x].isnull().sum()>0]
```

```python
# Mis_Str_var3 = train3[Mis_Str_var3].isnull().sum().sort_values(ascending = False).keys()
Mis_Str_var3
```

```python
Num_var_median = Mis_Num_var3
Str_var_mode = Mis_Str_var3[:8]
Str_var_missing = Mis_Str_var3[8:]
```

```python
Str_var_missing
```

```python
Num_var_median_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='median'))])
Str_var_mode_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='most_frequent'))])
Str_var_missing_imputer = Pipeline(steps = [('imputer', SimpleImputer(strategy='constant', fill_value = 'missing'))])
```

```python
preprocessor1 = ColumnTransformer(transformers=[('median', Num_var_median_imputer, Num_var_median)])
preprocessor2 = ColumnTransformer(transformers=[('mode', Str_var_mode_imputer, Str_var_mode),
                                                ('missing', Str_var_missing_imputer, Str_var_missing)])
```

```python
preprocessor2.named_transformers_['mode'].steps_name['imputer']
```

```python
Ntrain3 = preprocessor1.fit_transform(train3)
```

```python
Strain3 = preprocessor2.fit_transform(train3)
```

```python
Ntrain3 = pd.DataFrame(Ntrain3, columns=Num_var_median)
Strian3 = pd.DataFrame(Strain3, columns=Str_var_mode+Str_var_missing)
```
