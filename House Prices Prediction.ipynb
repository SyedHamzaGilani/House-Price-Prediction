# Data Cleaning With Different Strategies

## Strategy : 1
## Measuring Central Tendency

```python
Num_Var = train1.select_dtypes(include = ['int64', 'float64']).keys()
Str_Var = train1.select_dtypes(include = 'object').keys()
```

```python
Mis_Num_Var = [x for x in train1.select_dtypes(include = ['int64', 'float64']).keys() if train1[x].isnull().sum()>0]
Mis_Num_Var
```

```python
Mis_Str_Var = [x for x in train1.select_dtypes(include = 'object').keys() if train1[x].isnull().sum()>0]
Mis_Str_Var
```

```python
plt.figure(figsize = (16,9))
sns.set('poster')
for index, variable in enumerate(Mis_Num_Var):
    plt.subplot(2,2,index+1)
    sns.histplot(data = train1[variable], kde = True, line_kws = {'linewidth':10})
plt.show()
```

```python
Mis_Num_Var
```

```python
for index, variable in enumerate(Mis_Num_Var):
    if index == 0:
        train1[variable].replace(np.nan, train1[variable].mean(), inplace = True)
    else:
        train1[variable].replace(np.nan, train1[variable].median(), inplace = True)
```

```python
train1[Mis_Num_Var].isnull().sum()
```

```python
train1[Num_Var].isnull().sum()
```

```python
plt.figure(figsize = (16, 9))
sns.set('poster')
for index, variable in enumerate(Mis_Num_Var):
    plt.subplot(2,2,index+1)
    sns.histplot(data = train[variable], kde = True, line_kws = {'linewidth':10}, color='b')
    sns.histplot(data = train1[variable], kde = True, line_kws = {'linewidth':10}, color='r')
plt.show()
```

```python
Mis_Str_Var
```

```python
train1['MasVnrType'].mode()[0]
```

```python
for variable in Mis_Str_Var:
    train1[variable].replace(np.nan, train1[variable].mode()[0], inplace = True)
    print(f"{variable} Is Replaced With {train1[variable].mode()[0]}\n")
```
